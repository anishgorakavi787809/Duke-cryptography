This program counts how many instances of a letter 

```python
book = {}
#We ask user to input word and add each letter to or dictionary
word = input("enter word?:")

for w in word:
    try:
        if w == " ":
            continue
        book[w] += 1
    except:
        book[w] = 1

#We will print all the dictionary
for b in book:
    print(f"{b}:{book[b]}")
print("\n")
#We print mode
best = word[0]
best = book[best]
letterbest = ""
for b in book:
    if book[b] > best:
        best = book[b]
        letterbest = b
        continue
        
print(f"Most often: {letterbest}{best}")
print("\n")
#Calculating total
num = 0

for b in book:
    num += book[b]

percent_table = {}

for b in book:
    percent_table[b] = (book[b]/num) * 100

for p in percent_table:
    print(f"{p}:{float(percent_table[p])}%")
```

Bigram counter

```python
word = input("enter word?:")
book = {}
bicount = 0
word = word.replace(" ","")
print(word)
while bicount < len(word) - 1:
    try:
        book[word[bicount] + word[bicount + 1]] += 1
    except:
        book[word[bicount] + word[bicount + 1]] = 1
    bicount += 2
for b in book:
    print(b,book[b])

best = 0
be = ""
for b in book:
    if book[b] > best:
        best = book[b]
        be = b
print("Best is:",be,best)
```

```python
def coneztable(n,num):
    arr = []
    for i in range(n):
        ans = num * i
        if ans >= n:

            arr.append(ans % n)
        else:
            arr.append(ans)
    return arr


def createz(n):
    arr = []
    for i in range(n):
        arr.append(coneztable(n,i))
    return arr

def isField(arr):
    indx = 1
    field = False
    while indx < len(arr):
        
        if arr[indx] == 1:
            field = True 
        indx += 1
    return field
def checkifzfield(arr):
    
    for i in range(1,len(arr)):
        if not isField(arr[i]):
            return False
    return True
zs = ""
times = int(input("How many times n to run?:"))
for n in range(2,times + 1):
    print(createz(n))
    if checkifzfield(createz(n)):
        zs += str(n) +" "
print("n's that are field:",zs)
```
